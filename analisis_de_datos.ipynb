# Análisis de datos mediante Python

```python
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
user = 'root'
pw = 'RootSQL4703_'
host = 'localhost'
db = 'trafico_vehicular'

engine = create_engine(f'mysql+pymysql://{user}:{pw}@{host}/{db}')
```

```python
ciudades = pd.read_sql_table('ciudades',engine)
intercensal = pd.read_sql_table('intercensal',engine)
```

```python
ciudades.sort_values(by='indice_trafico',ascending=False).head(5)
```

```python
ciudades['indice_trafico'].plot(kind='hist', bins=10, edgecolor='black')
plt.title('Distribución del Índice de Trafico en México')
plt.xlabel('Índice de Trafico')
plt.ylabel('Número de Ciudades')
plt.show()
```

```python
sns.set_theme(style='ticks')

sns.barplot(data=ciudades.sort_values(by='indice_trafico',ascending=False).head(5), x='indice_trafico', y='estado', hue='indice_trafico', palette='dark:b_r')
plt.legend().remove()
plt.show()
```
